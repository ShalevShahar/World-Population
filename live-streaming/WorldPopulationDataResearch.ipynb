```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("WPP2019_TotalPopulationBySex.csv")
data['PopTotal'] = data['PopTotal'].astype(float)
```

```python
year2019= data["Time"]== 2019
year2020= data["Time"]== 2020
VariantMedium = data["Variant"]== 'Medium'
OnlyCountries = data['LocID']  < 895
data1 = pd.DataFrame(data[['LocID','Location','PopTotal']][VariantMedium & year2019 & OnlyCountries]).sort_values(by = 'LocID')
data1 = data1.reset_index()
data1['PopTotal'] = data1['PopTotal'] * 1000 
data1['PopTotal'] = data1['PopTotal'].astype(int)
data2['PopTotal'] = data2['PopTotal'] *1000
data2['PopTotal'] = data2['PopTotal'].astype(int)
data2 = data[['Location','PopTotal']][VariantMedium & year2020 & OnlyCountries]
```

```python
data2
```

```python
dataMerged = data1.merge(data2, left_on='Location', right_on='Location')
```

```python
dataMerged.rename(columns={"PopTotal_x": "pop2019", "PopTotal_y": "pop2020"})
dataMerged2 = dataMerged.drop(['index', 'LocID'], axis=1).rename(columns={"PopTotal_x": "pop2019", "PopTotal_y": "pop2020"})
```
